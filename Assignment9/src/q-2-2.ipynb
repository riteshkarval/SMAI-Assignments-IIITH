```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.mixture import GaussianMixture
from sklearn import metrics
from sklearn.decomposition import PCA
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

# PCA 16

```python
pca = PCA(n_components=16, whiten=False)
data = pca.fit_transform(digits.data)

n_components = np.arange(1, 16)
BIC = np.zeros(n_components.shape)
for i, n in enumerate(n_components):
    gmm = GaussianMixture(n_components=n)
    gmm.fit(data)
    BIC[i] = gmm.bic(data)
n_components = np.argmin(BIC)
print("No of components",n_components)
gmm = GaussianMixture(n_components=n_components)
gmm.fit(data)
sampled_points = gmm.sample(48)[0]
sampled_points = pca.inverse_transform(sampled_points)

sampled_points = sampled_points.reshape((48,8,8))
real_data = digits.data[:48].reshape((48,8,8))
print("Input Data")
plt.figure(figsize=(16,8))
columns = 5
for i, image in enumerate(real_data):
    plt.subplot(len(real_data) / columns + 1, columns, i + 1)
    plt.imshow(image)
```

```python
print("Density estimated points")
plt.figure(figsize=(16,8))
columns = 5
for i, image in enumerate(sampled_points):
    plt.subplot(len(real_data) / columns + 1, columns, i + 1)
    plt.imshow(image)
```
